```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def find_contours(img):
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur_img = cv2.GaussianBlur(gray_img, (13, 13), 0)
    ret, binary = cv2.threshold(blur_img, 80, 255, cv2.THRESH_BINARY_INV)
    binary_blur = cv2.medianBlur(binary, 21)
    contours, hierarchy = cv2.findContours(binary_blur, cv2.RETR_TREE,
                                           cv2.CHAIN_APPROX_SIMPLE)
    return contours

def draw_contours(img, contours):
    for i, c in enumerate(contours):
        area = cv2.contourArea(c)
        cv2.drawContours(img, contours, i, (0, 0, 255), 1)
    return img
```

```python
def bgr2gray(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
def get_magnitude(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    return 20*np.log(np.abs(fshift))
```

```python
def adjust_brightness(img, gamma):
    return np.power(img/float(np.max(img)), gamma)
```

```python
img = cv2.imread("anomal2.bmp", 0)
img2 = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
contours = find_contours(img2)
img2 = draw_contours(img2, contours)

cv2.imshow('frame', img2)
cv2.waitKey(0)
```

```python
cap = cv2.VideoCapture('test.avi')

# 以迴圈從影片檔案讀取影格，並顯示出來
while(cap.isOpened()):
    ret, frame = cap.read()
    contours = find_contours(frame)
    frame = draw_contours(frame, contours)
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
